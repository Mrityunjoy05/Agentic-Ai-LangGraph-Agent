```python
from diffusers import DiffusionPipeline
import torch
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0",
                                         torch_dtype=torch.float16,
                                         use_safetensors=True,
                                         variant="fp16")

pipe.to("cuda")
```

```python
negative_prompt = "cartoon, painting, illustration, blurry, deformed, ugly, low quality, CGI, toy, fake, animated, artificial background"
```

```python
prompt = (
    "dynamic full-body shot of Spider-Man mid-swing between skyscrapers, "
    "classic red and blue suit, web strands catching golden hour sunlight, "
    "New York City skyline background, motion blur on webs, "
    "dramatic upward angle, lens flare, photorealistic, cinematic, 8k"
)
```

```python
def generate_image(prompt, negative_prompt , steps = 50 , guidance = 7.5):
    print(f"Generating image for: {prompt}")

    image = pipe(
        prompt,
        negative_prompt=negative_prompt,
        height=1024,
        width=1024,
        num_inference_steps=int(steps),
        guidance_scale=float(guidance)
    ).images[0]

    return image
```

```python
import gradio as gr
# Launch Gradio
demo = gr.Interface(
    fn=generate_image,
    inputs=[
        gr.Textbox(label="Prompt", placeholder="a futuristic city at night, cyberpunk style"),
        gr.Textbox(label="Negative Prompt", value="cartoon, painting, illustration, blurry, deformed, ugly, low quality, CGI, toy, fake, animated, artificial background"),
        gr.Slider(20, 100, value=50, step=1, label="Steps"),
        gr.Slider(1, 20, value=7.5, step=0.5, label="Guidance Scale")
    ],
    outputs=gr.Image(label="Generated Image"),
    title="SDXL Image Generator",
    description="Free GPU image generation using Google Colab T4"
)

demo.launch(share=True)
```

```python
demo.close()
print("Server stopped ✅")
```
